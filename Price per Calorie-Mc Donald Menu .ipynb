## Question:

### Write a code to merge the two datasets, calculate the price per calorie of food, and stack rank the foods from the 'best value' (most calories per dollar*) to 'worst value' (least calories per dollar*).

The two given datasets:

Menu Item Prices:
https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/mcD_food_prices.csv

and 

Menu Item Calories:
https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/mcD_food_calories.csv

### Solution:

Let us import and preview the two datasets

```python
import pandas as pd
import numpy as np 
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
item_prices = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/mcD_food_prices.csv')
item_calories = pd.read_csv('https://raw.githubusercontent.com/erood/interviewqs.com_code_snippets/master/Datasets/mcD_food_calories.csv')
```

```python
item_prices.info()
item_calories.info()

item_prices.head(10)
item_calories.head(10)
```

### Merging both the datasets on "Food"

```python
items_merged = item_prices.merge(item_calories, how ='left', on = 'Food')
items_merged.head()

```

### Now, to calculate the calories per dollar - Create a new column with the calorie per dollar calculation

```python
items_merged['price_per_calorie']= items_merged['Price_USD']/items_merged['Calories']
items_merged
```

###  Sorting the Dataset obtained to Rank the Items from Most Value to Least Value

```python
items_merged = items_merged.sort_values(by='price_per_calorie', ascending=True)
```

Viewing the Head section of this dataset will give us the list of Least Value Food

```python
items_merged.head(10)
```

Viewing the Tail section of this dataset will give us the list of Best Value Food (most calories per dollar)
